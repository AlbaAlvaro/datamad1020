## 1. Obtain all the commits made in the past week via API, which is a JSON array that contains multiple commit objects.

```python
import requests
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
gh_token=os.getenv("GITHUB_TOKEN")
```

```python
url= "https://api.github.com"
headers={"Authorization": f"token {gh_token}"}
endpoint="/repos/{owner}/{repo}/commits"
repo_info={"owner":"ironhack-datalabs", "repo":"datamad1020"}
data_commits=requests.get(url+endpoint.format(**repo_info), headers=headers)
data_commits=data_commits.json()
print(data_commits)
```
